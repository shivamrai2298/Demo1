```python
def pqr(a, image_path=None):
    import smtplib
    from email.mime.text import MIMEText
    from email.mime.multipart import MIMEMultipart
    from email.mime.image import MIMEImage
    from datetime import datetime

    sender_email = "shivam.rai@impactguru.com"  
    app_password = "bxoh enuh nllt urem"    
    recipient_email = "shivamrai2298@gmail.com"  

    subject = "Message from Jupyter"
    message = a

    # Create a multipart message
    msg = MIMEMultipart()
    msg['Subject'] = subject
    msg['From'] = sender_email
    msg['To'] = recipient_email

    msg.attach(MIMEText(message, 'plain'))


    if image_path:
        try:
            with open(image_path, 'rb') as f:
                img_data = f.read()
            image = MIMEImage(img_data, name=image_path.split('/')[-1])
            msg.attach(image)
        except Exception as e:
            print(f"Error attaching image: {e}")


    try:
        server = smtplib.SMTP('smtp.gmail.com', 587)
        server.starttls()
        server.login(sender_email, app_password)
        server.send_message(msg)
        server.quit()
        print("k")
    except Exception as e:
        print("X")
```
